## o que é programação orientada a objetos (POO) ?



```python
class Pessoa:
    def __init__(self, nome,   idade):
        self.nome   =  nome
        self.idade  = idade

    def falar(self):
        print(f"{self.nome} está    falando.")

p1 = Pessoa("Ana", 22)
p2 = Pessoa("Carlos", 30)

p1.falar()
p2.falar()
```

```python
class Carro:
    def __init__(self, marca, modelo, ano):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano

    def ligar(self):
        print(f"O carro {self.marca} {self.modelo} foi ligado.")

carro1 = Carro("VW", "Jetta variant", 2012)
carro1.ligar()
```

```python
class ContaBancaria:
    def __init__(self):
        self._saldo = 0

    def get_saldo(self):
        return self._saldo

    def set_saldo(self, valor):
        if valor >= 0:
            self._saldo = valor
        else:
            print("Erro: saldo não pode ser negativo!")
```
